### 1. Explorando el API de usuarios de GitHub
Ahora vamos a explorar un nuevo API: el API de usuarios de GitHub. La URL de este API es https://api.github.com/users/{username}, donde {username} es el nombre del usuario en GitHub. Por ejemplo, aquí tenéis la URL para obtener información del usuario de Isra https://api.github.com/users/gootyfer. Si ponéis esta URL en una nueva pestaña del navegador podréis observar qué datos nos devuelve el API.

Tu tarea consiste en hacer una petición desde Python para obtener la información de tu propio usuario de GitHub. Después, almacena los datos recibidos en un diccionario y muestra el contenido por consola para comprobar que se ha cargado correctamente.

```python
import requests
import pandas as pd
```

```python
def llamar_api(endpoint):
    datos = requests.get(endpoint)
    print(f"llamar al api con respuesta : {datos.status_code}")
    
    if datos.status_code != 200:
        print(f"llamada falló por el siguiente motivo {datos.reason}")

    else:
        return datos.json()
   
```

```python
Datos_Jenn = llamar_api("https://api.github.com/users/jenniferamosmerino-gif")
```
